```python
class Profiler():
    
    # 1.
    def __init__(self, df, directory='EDA_'+date.today().strftime('%d-%m-%Y')) : 
        
        '''
        
        M√©todo constructor de la clase Profiler.
        
        Parameters
        ----------
        df : pandas.DataFrame
            Dataset de inter√©s para generar descripci√≥n autom√°tica
        directory : str
            Nombre de la carpeta de trabajo donde se almacenar√°n resultados y visualizaciones 
        
        '''
        
        if isinstance(df, pd.DataFrame):
            self.df = df
            self.directory = directory
            os.makedirs(self.directory, exist_ok=True)
        else:
            raise TypeError(
                'Debes entregar un DataFrame de Pandas como argumento del constructor de Profiler'
            )
    
    # 2.
    def summarize(self, include=None, select_dtypes=True, main_delimiter='?') :
        
        '''
        
        La funci√≥n caracteriza las variables del Dataset.
        
        Parameters
        ----------
        include : ‚Äòall‚Äô, lista de dtypes/columnas o None, default None
            Lista de tipos de datos/columnas para incluir en la salida de la funci√≥n
            - ‚Äòall‚Äô : Todas las columnas de la entrada se incluir√°n en la salida
            - Lista de dtypes/columnas : S√≥lo se incluyen los tipos de datos/columnas proporcionados 
            - None : S√≥lo se incluyen los tipos de datos numericos
        select_dtypes : bool, default True
            Booleano para especificar si la lista del parametro ‚Äòinclude‚Äô corresponde a tipos de datos (True) o
            corresponde a columnas espec√≠ficas del Dataset (False)
        main_delimiter : str, default ‚Äò?‚Äô
            Car√°cter separador para posibles columnas no at√≥micas en el Dataset
        
        Returns
        -------
        report : pandas.DataFrame
            Reporte con las principales caracter√≠sticas de las variables del Dataset
                
        '''
        
        # Implementar una funcionalidad para filtrar y aplicar este m√©todo a una o m√°s variables de inter√©s.
        if include==None :
            df_subset = self.df.select_dtypes(include=np.number).copy()
        elif include=='all' :
            df_subset = self.df.copy()
        else :
            if select_dtypes == True :
                df_subset = self.df.select_dtypes(include=include).copy()
            else :
                df_subset = self.df[include].copy()

        report = pd.DataFrame(index=df_subset.columns)
        
        # Reportar el tipo de variable
        report['dtypes'] = df_subset.dtypes.values
        
        # Reportar el n√∫mero y/o porcentaje de valores √∫nicos de la variable 
        report['uniques'] = [len(df_subset[col].unique()) for col in df_subset.columns]
        
        # Reportar el n√∫mero y/o porcentaje de valores nulos
        report['isna'] = df_subset.isna().sum().values
        report['isna_as_percentage'] = (report['isna'] / len(df_subset) * 100).round(2)
        
        # Reportar el n√∫mero y/o porcentaje de valores duplicados (ADICIONAL)
        report['duplicated'] = [df_subset[c].duplicated().sum() for c in df_subset.columns]
        report['duplicated_as_percentage'] = (report['duplicated'] / len(df_subset) * 100).round(2)
        
        # Si la variables es num√©rica:
        num_cols = df_subset.select_dtypes(include='number').columns
        if len(num_cols) > 0 :
            # - Reportar el n√∫mero y/o porcentaje de valores cero, negativos y outliers
            report['zeros'] = np.nan
            report.loc[num_cols, 'zeros'] = [(df_subset[col]==0).sum() for col in num_cols]
            report['zeros_as_percentage'] = (report['zeros'] / len(df_subset) * 100).round(2)
            report['negatives'] = np.nan
            report.loc[num_cols, 'negatives'] = [(df_subset[col]<0).sum() for col in num_cols]
            report['negatives_as_percentage'] = (report['negatives'] / len(df_subset) * 100).round(2)
            report['outliers'] = np.nan
            report.loc[num_cols, 'outliers'] = [(np.abs(stats.zscore(df_subset[col]))<3).sum() for col in num_cols]
            report['outliers_as_percentage'] = (report['outliers'] / len(df_subset) * 100).round(2)
            # - Reportar estad√≠stica descriptiva como el valor m√≠nimo, m√°ximo, promedio y los percentiles 25, 50, 75 y 100
            report = pd.concat([report,df_subset.describe().T.drop('count', axis=1)], axis=1)
        
        # Levantar una alerta en caso de encontrar alguna anomal√≠a fuera de lo com√∫n 
        # (el criterio debe ser ajustable por el usuario)
        anomaly_cols = []
        for c in df_subset.select_dtypes('O').columns : 
            if len(df_subset[df_subset[c].str.contains(main_delimiter, na=False, regex=False)]) > 0 :
                anomaly_cols.append(c)
        if len(anomaly_cols) > 0:
            warnings.warn(
                'Una/s de la/s columna/s del Dataset presenta datos *no at√≥micos*.', UserWarning
            )
        
        # Guardar sus resultados en el directorio `EDA_fecha/summary.txt`. 
        # El archivo debe separar de forma clara y ordenada los resultados de cada punto.
        report = report.fillna('---')
        report_txt = ''
        report_txt += '- Tipos de variables :'.upper() + '\n\n'
        cols = ['dtypes']
        report_txt += str(report[cols]) + '\n\n'
        report_txt += '- N√∫mero y/o porcentaje de valores √∫nicos de la variable :'.upper() + '\n\n'
        cols = ['uniques']
        report_txt +=  str(report[cols]) + '\n\n'
        report_txt += '- N√∫mero y/o porcentaje de valores nulos :'.upper() + '\n\n'
        cols = ['isna', 'isna_as_percentage']
        report_txt +=  str(report[cols]) + '\n\n'
        report_txt += '- N√∫mero y/o porcentaje de valores duplicados :'.upper() + '\n\n'
        cols = ['duplicated', 'duplicated_as_percentage']
        report_txt +=  str(report[cols]) + '\n\n'
        if len(num_cols) > 0 :
            report_txt += '- N√∫mero y/o porcentaje de valores cero, negativos y outliers :'.upper() + '\n\n'
            cols = ['zeros', 'zeros_as_percentage', 'negatives', 'negatives_as_percentage', 
                    'outliers', 'outliers_as_percentage']
            report_txt +=  str(report[cols]) + '\n\n'
            report_txt += '- Estad√≠stica descriptiva :'.upper() + '\n\n'
            cols = ['mean', 'std', 'min', '25%', '50%', '75%', 'max']
        report_txt +=  str(report[cols]) + '\n\n'
        report_file = open(self.directory+'/summary.txt', "w")
        report_file.write(report_txt)
        report_file.close()
        return report

    # 3.
    def plot_vars(self, include=None, select_dtypes=True, n_samples=1000, random_state=42, n_categories=5) : 
        
        '''
        
        La funci√≥n grafica la distribuci√≥n e interraciones de las variables del Dataset. 
        
        Parameters
        ----------
        include : ‚Äòall‚Äô, lista de dtypes/columnas o None, default None
            Lista de tipos de datos/columnas para incluir en la salida de la funci√≥n
            - ‚Äòall‚Äô : Todas las columnas de la entrada se incluir√°n en la salida
            - Lista de dtypes/columnas : S√≥lo se incluyen los tipos de datos/columnas proporcionados 
            - None : S√≥lo se incluyen los tipos de datos numericos
        select_dtypes : bool, default True
            Booleano para especificar si la lista del parametro ‚Äòinclude‚Äô corresponde a tipos de datos (True) o
            corresponde a columnas espec√≠ficas del Dataset (False)
        n_samples : int, default 1000
            Tama√±o de muestreo desde el Dataset para agilizar el proceso de generaci√≥n de visualizaciones
        random_state : int, default 42
            Semilla para inicializar el generador interno de n√∫meros aleatorios
        n_categories : int, default 5
            Cantidad de clases para mostrar en las visualizaciones de variables categoricas
            
        '''
        
        # Crear la carpeta `EDA_fecha/plots`
        folder = self.directory + '/plots'
        os.makedirs(folder, exist_ok=True)

        # Implementar una funcionalidad para filtrar y aplicar este m√©todo a una o m√°s variables de inter√©s.
        if select_dtypes == True :
            if include==None :
                df_subset = self.df.select_dtypes(include=np.number).copy()
            elif include=='all' :
                df_subset = self.df.copy()
            else : 
                df_subset = self.df.select_dtypes(include=include).copy()
        else :
            if include=='all' :
                df_subset = self.df.copy()
            else :
                df_subset = self.df[include].copy()
        
        # Para las variables num√©ricas:
        # - Genere un gr√°fico de distribuci√≥n de densidad
        # - Grafique la correlaci√≥n entre las variables
        # Para las variables categ√≥ricas:
        # - Genere un histograma de las top N categor√≠as (N debe ser un par√°metro ajustable)
        # - Grafique el coeficiente V de Cramer entre las variables
        # Guardar cada gr√°fico generado en la carpeta `EDA_fecha/plots` en formato `.pdf` y 
        # bajo el naming `variable.pdf`, donde `variable` es el nombre de la variable de inter√©s
        for col in df_subset.columns :
            serie = df_subset[col].copy()
            sampled_serie = pd.DataFrame(serie.sample(n_samples, random_state=random_state))
            if is_numeric_dtype(serie) or is_datetime64_any_dtype(serie) :
                fig, ax = plt.subplots(figsize=(20, 5), dpi=100)
                sns.kdeplot(data=sampled_serie, x=col, ax=ax, fill=True, cut=0, color='#abcae4', edgecolor='black')
                ax.set_title(f'{col} Profile'.upper(), 
                             fontsize=20, fontweight='bold', color='red')
                ax.set_xlabel(col, fontsize=15, fontweight='bold')
                ax.set_ylabel('Density', fontsize=15, fontweight='bold')
                ax.tick_params(labelsize=15)
            else :
                count = (
                    serie.value_counts()[0:n_categories]
                    .reset_index()
                    .rename(columns = {serie.name: 'Count'})
                )
                fig, ax = plt.subplots(figsize=(20, 5), dpi=100)
                sns.barplot(data=count, x='Count', y='index', ax=ax, color='#abcae4', edgecolor='black')
                ax.set_title(f'{n_categories} Most common categories of {serie.name}'.upper(), 
                             fontsize=20, fontweight='bold', color='red')
                ax.set_xlabel(col, fontsize=15, fontweight='bold')
                ax.set_ylabel('Count', fontsize=15, fontweight='bold')
                ax.tick_params(labelsize=15)
            plt.show() ;
            fig.savefig(folder+f"/{col}.pdf", dpi=100)
        sampled_df_subset = df_subset.sample(n_samples, random_state=random_state)
        if len(df_subset.select_dtypes(include=np.number).columns) > 0 :
            corr_mat = sampled_df_subset.dropna().corr(numeric_only=True)
            fig, ax = plt.subplots(figsize=(10, 10), dpi=100)
            ax = sns.heatmap(corr_mat, cmap='coolwarm', annot=True, fmt='.1%',
                             linewidths=.5, vmin=-1, vmax=1)
            ax.tick_params(labelsize=15)
            plt.xticks(rotation=90) 
            plt.yticks(rotation=0) 
            cbar_ax = ax.collections[0].colorbar
            cbar_ax.ax.tick_params(labelsize=15)
            cbar_ax.set_ticks([-1, -.5, 0, +.5, +1])
            cbar_ax.set_ticklabels(['-100 %', '-50 %', '0 %', '+50 %', '+100 %'])
            cbar_ax.set_label('Pearson Correlation Coefficient', rotation=90, fontsize=15, weight='bold')
            plt.show() ; 
            fig.savefig(folder+f"/CorrelationMatrix.pdf", dpi=100)    
        if len(df_subset.select_dtypes(include='O').columns) > 0 :
            def cramers_corrected_stat(confusion_matrix):
                """ calculate Cramers V statistic for categorical-categorical association.
                    uses correction from Bergsma and Wicher, 
                    Journal of the Korean Statistical Society 42 (2013): 323-328
                """
                chi2 = stats.chi2_contingency(confusion_matrix)[0]
                n = confusion_matrix.sum().sum()
                phi2 = chi2/n
                r,k = confusion_matrix.shape
                phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))    
                rcorr = r - ((r-1)**2)/(n-1)
                kcorr = k - ((k-1)**2)/(n-1)
                return np.sqrt(phi2corr / min( (kcorr-1), (rcorr-1)))
            cols = df_subset.select_dtypes(include='O').columns.tolist()
            corrM = np.zeros((len(cols),len(cols)))
            for col1, col2 in itertools.combinations(cols, 2):
                idx1, idx2 = cols.index(col1), cols.index(col2)
                corrM[idx1, idx2] = cramers_corrected_stat(pd.crosstab(sampled_df_subset[col1], sampled_df_subset[col2]))
                corrM[idx2, idx1] = corrM[idx1, idx2]
            corr_mat = pd.DataFrame(corrM, index=cols, columns=cols)
            fig, ax = plt.subplots(figsize=(10, 10), dpi=100)
            ax = sns.heatmap(corr_mat, cmap='coolwarm', annot=True, fmt='.1%',
                             linewidths=.5, vmin=0, vmax=1)
            ax.tick_params(labelsize=15)
            plt.xticks(rotation=90) 
            plt.yticks(rotation=0) 
            cbar_ax = ax.collections[0].colorbar
            cbar_ax.ax.tick_params(labelsize=15)
            cbar_ax.set_ticks([0, .25, .5, .75, 1])
            cbar_ax.set_ticklabels(['0 %', '25 %', '50 %', '75 %', '100 %'])
            cbar_ax.set_label("Cramer's V Coefficient", rotation=90, fontsize=15, weight='bold')
            plt.show() ; 
            fig.savefig(folder+f"/CramerVMatrix.pdf", dpi=100)    
    
    # 4.
    def clean_data(self, include=None, select_dtypes=True, main_delimiter='?', strategy='drop') :
        
        '''
        
        La funci√≥n limpia los datos del Dataset para que luego puedan ser procesados
        
        Parameters
        ----------
        include : ‚Äòall‚Äô, lista de dtypes/columnas o None, default None
            Lista de tipos de datos/columnas para incluir en la salida de la funci√≥n
            - ‚Äòall‚Äô : Todas las columnas de la entrada se incluir√°n en la salida
            - Lista de dtypes/columnas : S√≥lo se incluyen los tipos de datos/columnas proporcionados 
            - None : S√≥lo se incluyen los tipos de datos numericos
        select_dtypes : bool, default True
            Booleano para especificar si la lista del parametro ‚Äòinclude‚Äô corresponde a tipos de datos (True) o
            corresponde a columnas espec√≠ficas del Dataset (False)
        main_delimiter : str, default ‚Äò?‚Äô
            Car√°cter separador para posibles columnas no at√≥micas en el Dataset
        strategy: str, default ‚Äòdrop‚Äô
            Estrategia de imputacion para tratar los valores nulos
            - ‚Äòdrop‚Äô : Eliminar valores nulos
            - ‚Äòsimple_imputer‚Äô : Reemplazar valores nulos usando la media/moda de las variables num√©ricas/categ√≥ricas
            
        Returns
        -------
        df_subset_clean : pandas.DataFrame
            Dataset limpiado seg√∫n especificaciones
            
        '''
        
        # Crear la carpeta `EDA_fecha/clean_data`
        folder = self.directory + '/clean_data'
        os.makedirs(folder, exist_ok=True)
        
        # Implementar una funcionalidad para filtrar y aplicar este m√©todo a una o m√°s variables de inter√©s.
        if select_dtypes == True :
            if include==None :
                df_subset = self.df.select_dtypes(include=np.number).copy()
            elif include=='all' :
                df_subset = self.df.copy()
            else : 
                df_subset = self.df.select_dtypes(include=include).copy()
        else :
            if include=='all' :
                df_subset = self.df.copy()
            else :
                df_subset = self.df[include].copy()
        
        df_subset_clean = df_subset.copy()

        # Drop de valores duplicados
        df_subset_clean.drop_duplicates(inplace=True)
                
        # Una de las columnas del dataframe presenta datos *no at√≥micos*. Separe dicha columna en las columnas que la compongan.
        # - Hint: ¬øQu√© caracteres permiten separar una columna de otra?
        # R. Los caracteres que permiten separar la columna son '?', ':', '*' y '(', siendo '?' el caracter principal
        # - Para las pruebas con el dataset nuevo, puede esperar que exista al menos una columna con este tipo de problema. 
        # Asuma que los separadores ser√°n los mismos, aunque el n√∫mero de columnas a separar puede ser distinto.
        # Deber√≠an usar `FunctionTransformer`.
        anomaly_cols = []
        for c in df_subset.select_dtypes('O').columns : 
            if len((df_subset_clean[df_subset_clean[c].str.contains(main_delimiter, na=False, regex=False)])) > 0 :
                anomaly_cols.append(c)
        if len(anomaly_cols) > 0:
            def anomaly_separator(series_in) :
                series_out = series_in.copy()
                series_out = series_out.str.split('[?\:\*\(]', expand=True)
                series_out.columns = [c.capitalize() for c in series_in.name.split(detect(series_in.name))]
                return series_out
            transformer = FunctionTransformer(anomaly_separator)
            for col in anomaly_cols :
                df_subset_clean = pd.concat([df_subset_clean, transformer.transform(df_subset_clean[col])], axis=1)
                df_subset_clean = df_subset_clean.drop(col, axis=1)
        df_subset_clean = df_subset_clean.replace('nan', np.nan, regex=True).astype('float', errors='ignore')
        df_subset_clean_cols = df_subset_clean.columns
                
        # Implementar como m√≠nimo 2 t√©cnicas para tratar los valores nulos, como:
        # - Drop de valores nulos
        # - Imputar valores nulos con alguna t√©cnica de imputaci√≥n
        # - Funcionalidad para escoger entre una t√©cnica y la otra.
        num_cols = df_subset_clean.select_dtypes(include=np.number).columns.tolist()
        cat_cols = [col for col in df_subset_clean.columns if col not in num_cols]
        if strategy=='drop' : 
            df_subset_clean = df_subset_clean.dropna()
        elif strategy=='simple_imputer' :
            ct = ColumnTransformer(
                [('trf1', SimpleImputer(strategy='mean'), num_cols),
                 ('trf2', SimpleImputer(strategy='most_frequent'), cat_cols)
                ], remainder='drop')
            df_subset_clean = pd.DataFrame(ct.fit_transform(df_subset_clean), columns=num_cols+cat_cols)
            df_subset_clean = df_subset_clean.reindex(df_subset_clean_cols, axis=1)
        df_subset_clean = df_subset_clean.reset_index(drop=True)
        
        # Guardar los datos procesados en formato `.csv` en el path `EDA_fecha/clean_data/data.csv`
        df_subset_clean.to_csv(folder + '/data.csv', sep=';', index=False)
        return df_subset_clean
        
    # 5.
    def scale(self, include=None, select_dtypes=True, main_delimiter='?', strategy='drop', 
              cat_preprocess=[OneHotEncoder], num_preprocess=[np.log10, MinMaxScaler]) :
        
        '''
        
        La funci√≥n prepara adecuadamente los datos para luego ser consumidos por alg√∫n tipo de algoritmo. 
        
        Parameters
        ----------
        include : ‚Äòall‚Äô, lista de dtypes/columnas o None, default None
            Lista de tipos de datos/columnas para incluir en la salida de la funci√≥n
            - ‚Äòall‚Äô : Todas las columnas de la entrada se incluir√°n en la salida
            - Lista de dtypes/columnas : S√≥lo se incluyen los tipos de datos/columnas proporcionados 
            - None : S√≥lo se incluyen los tipos de datos numericos
        select_dtypes : bool, default True
            Booleano para especificar si la lista del parametro ‚Äòinclude‚Äô corresponde a tipos de datos (True) o
            corresponde a columnas espec√≠ficas del Dataset (False)
        main_delimiter : str, default ‚Äò?‚Äô
            Car√°cter separador para posibles columnas no at√≥micas en el Dataset
        strategy: str, default ‚Äòdrop‚Äô
            Estrategia de imputacion para tratar los valores nulos
            - ‚Äòdrop‚Äô : Eliminar valores nulos
            - ‚Äòsimple_imputer‚Äô : Reemplazar valores nulos usando la media/moda de las variables num√©ricas/categ√≥ricas
        cat_preprocess: list, default [OneHotEncoder]
            T√©cnica de escalamiento para las variables categ√≥ricas
        num_preprocess: list, default [np.log10, MinMaxScaler]
            T√©cnica de escalamiento para las variables num√©ricas
            
        Returns
        -------
        df_subset_scale : pandas.DataFrame
            Dataset escalado seg√∫n especificaciones
        
        '''
        
        # Crear la carpeta EDA_fecha/scale
        folder = self.directory + '/scale'
        os.makedirs(folder, exist_ok=True)
        
        df_subset_scale = Profiler(self.df).clean_data(include, select_dtypes, main_delimiter, strategy)
        num_cols = df_subset_scale.select_dtypes(include=np.number).columns.tolist()
        cat_cols = [col for col in df_subset_scale.columns if col not in num_cols]
        
        # Procesar de forma adecuada los datos num√©ricos y categ√≥ricos:
        # - Su m√©todo debe recibir las t√©cnicas de escalamiento como argumento de entrada 
        # (utilizar solo t√©cnicas compatibles con el framework de `sklearn`)
        # - Para los atributos num√©ricos, se transforme los datos con un escalador logar√≠tmico y un `MinMaxScaler`
        # - Asuma que no existen datos ordinales en su dataset
        # Guardar todo este procesamiento en un `ColumnTransformer`.
        preprocessor = ColumnTransformer(
            transformers=[
                ('cat_preprocess', cat_preprocess[0](), cat_cols),
                ('num_preprocess', Pipeline(
                    steps=[("num_preprocess1", FunctionTransformer(num_preprocess[0])),
                           ('num_preprocess2', num_preprocess[1]())
                          ]), num_cols)
            ], remainder='drop'
        )
        df_subset_scale = pd.DataFrame(preprocessor.fit_transform(df_subset_scale))

        # Guardar los datos limpios y transformados en formato `.csv` en el path `EDA_fecha/process/scaled_features.csv`
        df_subset_scale.to_csv(folder + '/scaled_features.csv', sep=';', index=False)
        return df_subset_scale
    
    # 6.
    def make_clusters(self, include=None, select_dtypes=True, main_delimiter='?', strategy='drop', 
                      cat_preprocess=[OneHotEncoder], num_preprocess=[np.log10, MinMaxScaler],
                      random_state=42, make_clusters_method=[KMeans], n_clusters=None, threshold=.05) :
        
        '''
        
        La funci√≥n genera clusters de los datos usando alg√∫n algoritmo de clusterizaci√≥n.
        
        Parameters
        ----------
        include : ‚Äòall‚Äô, lista de dtypes/columnas o None, default None
            Lista de tipos de datos/columnas para incluir en la salida de la funci√≥n
            - ‚Äòall‚Äô : Todas las columnas de la entrada se incluir√°n en la salida
            - Lista de dtypes/columnas : S√≥lo se incluyen los tipos de datos/columnas proporcionados 
            - None : S√≥lo se incluyen los tipos de datos numericos
        select_dtypes : bool, default True
            Booleano para especificar si la lista del parametro ‚Äòinclude‚Äô corresponde a tipos de datos (True) o
            corresponde a columnas espec√≠ficas del Dataset (False)
        main_delimiter : str, default ‚Äò?‚Äô
            Car√°cter separador para posibles columnas no at√≥micas en el Dataset
        strategy: str, default ‚Äòdrop‚Äô
            Estrategia de imputacion para tratar los valores nulos
            - ‚Äòdrop‚Äô : Eliminar valores nulos
            - ‚Äòsimple_imputer‚Äô : Reemplazar valores nulos usando la media/moda de las variables num√©ricas/categ√≥ricas
        cat_preprocess: list, default [OneHotEncoder]
            T√©cnica de escalamiento para las variables categ√≥ricas
        num_preprocess: list, default [np.log10, MinMaxScaler]
            T√©cnica de escalamiento para las variables num√©ricas
        random_state : int, default 42
            Semilla para inicializar el generador interno de n√∫meros aleatorios
        make_clusters_method : list, default [KMeans]
            Algoritmo de clustering
        n_clusters : int o None, default None
            Cantidad de clusters para agrupar el Dataset
        threshold : float, default 0.05
            Valor de diferencia relativa m√≠nimo para definir ubicaci√≥n del codo

        Returns
        -------
        df_subset : pandas.DataFrame
            Dataset limpiado y agrupado seg√∫n especificaciones
        
        '''
        
        # Crear la carpeta `EDA_fecha/clusters`
        folder = self.directory + '/clusters'
        os.makedirs(folder, exist_ok=True)
        
        # Generar un estudio del codo donde se√±ale la cantidad de clusters optimos para el desarrollo.
        # Su m√©todo debe recibir el algoritmo de clustering como argumento de entrada 
        # (utilizar solo algoritmos compatibles con el framework de `sklearn`).
        # No olvide pre procesar adecuadamente los datos antes de implementar la t√©cnica de clustering. 
        # En este punto es espera que generen un `Pipeline` de sklearn. 
        # Adem√°s, su m√©todo deber√≠a usar lo construido en los puntos 4 y 5. 
        # Su m√©todo debe ser capaz de funcionar a partir de datos crudos (se descontar√° puntaje de lo contrario).
        
        df_subset = Profiler(self.df).scale(include, select_dtypes, main_delimiter, strategy, cat_preprocess, num_preprocess)
        
        SSE = []

        for k in range(1, 21): 
            pipe = Pipeline([('trf', make_clusters_method[0](n_clusters=k, random_state=random_state))])
            pipe.fit(df_subset)
            SSE.append(pipe.named_steps['trf'].inertia_)
        
        n_clusters_optim = abs(pd.Series(SSE).diff()/SSE[0]).lt(threshold).idxmax()
        
        fig, ax = plt.subplots(figsize=(20, 5), dpi=100)
        ax.plot(np.arange(1,21), SSE, 'bo-')
        ax.plot(n_clusters_optim, SSE[n_clusters_optim-1], 'ro', markersize=10, markeredgecolor='red')
        ax.axvline(n_clusters_optim, linestyle='--', color='red')
        ax.set_title(f'M√©todo del Codo\nCantidad √ìptima de Clusters = {n_clusters_optim}'.upper(), 
                     fontsize=20, fontweight='bold', color='red')
        ax.set_xlabel('N√∫mero de Clusters ', fontsize=15, fontweight='bold')
        ax.set_ylabel('SSE', fontsize=15, fontweight='bold')
        ax.set_xticks(np.arange(1, 21,1))
        ax.tick_params(labelsize=15)
        plt.show() ;
        fig.savefig(folder+f"/ElbowMethod.pdf", dpi=100)
        
        if n_clusters == None :
            n_clusters = n_clusters_optim
            
        pipe = Pipeline([('trf', make_clusters_method[0](n_clusters=n_clusters_optim, random_state=random_state))])
        pipe.fit(df_subset).transform(df_subset)
        clusters = pipe.named_steps['trf'].labels_ + 1
        
        # Una vez generado los clusters, proyecte los datos a 2 dimensiones usando su t√©cnica de reducci√≥n de 
        # dimensionalidad favorita y grafique los resultados coloreando por cluster.
        pipe = Pipeline([('trf', TSNE(n_components=2, random_state=random_state))])
        rates_tsne = pipe.fit(df_subset).fit_transform(df_subset)
        fig, ax = plt.subplots(figsize=(6, 6), dpi=100)
        sns.scatterplot(x=rates_tsne[:,0], y=rates_tsne[:,1], hue=clusters, ax=ax, palette='Set1', edgecolor='black')
        ax.set_title(f'Proyecci√≥n tSNE de Clusters'.upper(), fontsize=20, fontweight='bold', color='red')
        ax.set_xlabel('1ra Componente', fontsize=15, fontweight='bold')
        ax.set_ylabel('2da Componente', fontsize=15, fontweight='bold')
        ax.tick_params(labelsize=15)
        plt.show() ;
        fig.savefig(folder+f"/Clusters2D.pdf", dpi=100)

        # Guardar los datos con su respectivo cluster en formato `.csv` en el path `EDA_fecha/clusters/data_clusters.csv`. 
        # Guarde tambi√©n los gr√°ficos generados en el mismo path.
        df_subset = Profiler(self.df).clean_data(include, select_dtypes, main_delimiter, strategy)
        df_subset['Cluster'] = clusters
        df_subset.reset_index(drop=True).to_csv(folder + '/data_clusters.csv', sep=';', index=False)
        return df_subset

    # 7.
    def detect_anomalies(self, include=None, select_dtypes=True, main_delimiter='?', strategy='drop', 
                         cat_preprocess=[OneHotEncoder], num_preprocess=[np.log10, MinMaxScaler],
                         random_state=42, detect_anomalies_method=[DBSCAN]) :
        
        '''
        
        La funci√≥n detecta anomal√≠as en los datos.
        
        Parameters
        ----------
        include : ‚Äòall‚Äô, lista de dtypes/columnas o None, default None
            Lista de tipos de datos/columnas para incluir en la salida de la funci√≥n
            - ‚Äòall‚Äô : Todas las columnas de la entrada se incluir√°n en la salida
            - Lista de dtypes/columnas : S√≥lo se incluyen los tipos de datos/columnas proporcionados 
            - None : S√≥lo se incluyen los tipos de datos numericos
        select_dtypes : bool, default True
            Booleano para especificar si la lista del parametro ‚Äòinclude‚Äô corresponde a tipos de datos (True) o
            corresponde a columnas espec√≠ficas del Dataset (False)
        main_delimiter : str, default ‚Äò?‚Äô
            Car√°cter separador para posibles columnas no at√≥micas en el Dataset
        strategy : str, default ‚Äòdrop‚Äô
            Estrategia de imputacion para tratar los valores nulos
            - ‚Äòdrop‚Äô : Eliminar valores nulos
            - ‚Äòsimple_imputer‚Äô : Reemplazar valores nulos usando la media/moda de las variables num√©ricas/categ√≥ricas
        cat_preprocess : list, default [OneHotEncoder]
            T√©cnica de escalamiento para las variables categ√≥ricas
        num_preprocess : list, default [np.log10, MinMaxScaler]
            T√©cnica de escalamiento para las variables num√©ricas
        random_state : int, default 42
            Semilla para inicializar el generador interno de n√∫meros aleatorios
        detect_anomalies_method : list, default [DBSCAN]
            Algoritmo de detecci√≥n de anomal√≠as

        Returns
        -------
        df_subset : pandas.DataFrame
            Dataset limpiado y agrupado seg√∫n especificaciones

        '''
        
        # Crear la carpeta `EDA_fecha/anomalies`
        folder = self.directory + '/anomalies'
        os.makedirs(folder, exist_ok=True)
        
        # Implementar alguna t√©cnica de detecci√≥n de anomal√≠as.
        # Al igual que el punto anterior, su m√©todo debe considerar los siguientes puntos:
        # - No olvide pre procesar de forma adecuada los datos antes de implementar la t√©cnica de detecci√≥n de anomal√≠a. 
        # - En este punto es espera que generen un `Pipeline` de sklearn. Adem√°s, su m√©todo deber√≠a usar lo construido 
        # en los puntos 4 y 5. 
        # - Su m√©todo debe ser capaz de funcionar a partir de datos crudos (se descontar√° puntaje de lo contrario).
        # - Su m√©todo debe recibir el algoritmo como argumento de entrada
        
        df_subset = Profiler(self.df).scale(include, select_dtypes, main_delimiter, strategy, cat_preprocess, num_preprocess)
        
        pipe = Pipeline([('trf', detect_anomalies_method[0]())])
        pipe.fit(df_subset)
        anomalies = np.where(pipe.named_steps['trf'].labels_==-1, 'outlier', 'non-outlier')
        
        # - Una vez generado las etiquetas, proyecte los datos a 2 dimensiones y grafique los resultados coloreando por las 
        # etiquetas predichas por el detector de anomal√≠as
        pipe = Pipeline([('trf', TSNE(n_components=2, random_state=random_state))])
        rates_tsne = pipe.fit(df_subset).fit_transform(df_subset)
        fig, ax = plt.subplots(figsize=(6, 6), dpi=100)
        sns.scatterplot(x=rates_tsne[:,0], y=rates_tsne[:,1], hue=anomalies, ax=ax, palette='Set1', edgecolor='black')
        ax.set_title(f'Proyecci√≥n tSNE de Anomal√≠as'.upper(), fontsize=20, fontweight='bold', color='red')
        ax.set_xlabel('1ra Componente', fontsize=15, fontweight='bold')
        ax.set_ylabel('2da Componente', fontsize=15, fontweight='bold')
        ax.tick_params(labelsize=15)
        plt.show() ;
        fig.savefig(folder+f"/Anomalies2D.pdf", dpi=100)
        
        # - Guardar los datos con su respectiva etiqueta en formato `.csv` en el path `EDA_fecha/anomalies/data_anomalies.csv`. 
        # Guarde tambi√©n los gr√°ficos generados en el mismo path
        df_subset = Profiler(self.df).clean_data(include, select_dtypes, main_delimiter, strategy)
        df_subset['Anomaly'] = anomalies
        df_subset.reset_index(drop=True).to_csv(folder + '/data_clusters.csv', sep=';', index=False)
        return df_subset
    
    # 8.
    def profile(self, include=None, select_dtypes=True, main_delimiter='?', strategy='drop',  
                n_samples=1000, random_state=42, n_categories=5,
                cat_preprocess=[OneHotEncoder], num_preprocess=[np.log10, MinMaxScaler], 
                make_clusters_method=[KMeans], n_clusters=None, threshold=.05, 
                detect_anomalies_method=[DBSCAN]) :
        
        '''
        
        La funci√≥n ejecuta todos los m√©todos anteriores.
        
        Parameters
        ----------
        
        include : ‚Äòall‚Äô, lista de dtypes/columnas o None, default None
            Lista de tipos de datos/columnas para incluir en la salida de la funci√≥n
            - ‚Äòall‚Äô : Todas las columnas de la entrada se incluir√°n en la salida
            - Lista de dtypes/columnas : S√≥lo se incluyen los tipos de datos/columnas proporcionados 
            - None : S√≥lo se incluyen los tipos de datos numericos
        select_dtypes : bool, default True
            Booleano para especificar si la lista del parametro ‚Äòinclude‚Äô corresponde a tipos de datos (True) o
            corresponde a columnas espec√≠ficas del Dataset (False)
        main_delimiter : str, default ‚Äò?‚Äô
            Car√°cter separador para posibles columnas no at√≥micas en el Dataset
        strategy: str, default ‚Äòdrop‚Äô
            Estrategia de imputacion para tratar los valores nulos
            - ‚Äòdrop‚Äô : Eliminar valores nulos
            - ‚Äòsimple_imputer‚Äô : Reemplazar valores nulos usando la media/moda de las variables num√©ricas/categ√≥ricas
        n_samples : int, default 1000
            Tama√±o de muestreo desde el Dataset para agilizar el proceso de generaci√≥n de visualizaciones
        random_state : int, default 42
            Semilla para inicializar el generador interno de n√∫meros aleatorios
        n_categories : int, default 5
            Cantidad de clases para mostrar en las visualizaciones de variables categoricas
        cat_preprocess: list, default [OneHotEncoder]
            T√©cnica de escalamiento para las variables categ√≥ricas
        num_preprocess: list, default [np.log10, MinMaxScaler]
            T√©cnica de escalamiento para las variables num√©ricas
        make_clusters_method : list, default [KMeans]
            Algoritmo de clustering
        n_clusters : int o None, default None
            Cantidad de clusters para agrupar el Dataset
        threshold : float, default 0.05
            Valor de diferencia relativa m√≠nimo para definir ubicaci√≥n del codo
        detect_anomalies_method : list, default [DBSCAN]
            Algoritmo de detecci√≥n de anomal√≠as
        
        '''
        
        prof = Profiler(self.df)
        prof.summarize(include, select_dtypes, main_delimiter)
        prof.plot_vars(include, select_dtypes, n_samples, random_state, n_categories)
        
        option = input('¬øLimpiar datos y volver a graficar? [y]es, [n]o:\n')
        if option == 'y' :
            prof.clearGarbage()
            prof = Profiler(Profiler(self.df).clean_data(include, select_dtypes, main_delimiter, strategy))
            prof.summarize('all', main_delimiter)
            prof.plot_vars('all', n_samples, random_state, n_categories)
            
        option = input('¬øVisualizar clusters de agrupaci√≥n de datos? [y]es, [n]o:\n')
        if option == 'y' : 
            prof = Profiler(self.df)
            prof.make_clusters(include, select_dtypes, main_delimiter, strategy, cat_preprocess, num_preprocess,
                               random_state, make_clusters_method, n_clusters, threshold)
            
        option = input('¬øVisualizar anomal√≠as de datos? [y]es, [n]o:\n')
        if option == 'y' : 
            prof = Profiler(self.df)
            prof.detect_anomalies(include, select_dtypes, main_delimiter, strategy, cat_preprocess, num_preprocess,
                                  random_state, detect_anomalies_method)
            
        option = input('¬øEliminar carpeta de trabajo? [y]es, [n]o:\n')
        if option == 'y' : 
            prof.clearGarbage()
            
    # 9.
    def clearGarbage(self) :
        
        '''
        
        La funci√≥n elimina las carpetas/archivos creados/as por la clase Profiler.
            
        '''

        shutil.rmtree(self.directory) 
```

El resto de los atributos son descartados puesto que se ha supuesto, *a priori*, que no aportar√≠an informaci√≥n relevante en la agrupaci√≥n de los atletas.

Es importante se√±alar que las implicancias del supuesto enunciado son discutidas m√°s adelante en el an√°lisis de los resultados obtenidos a partir del proceso de creaci√≥n de clusters. Asimismo, se debe notar que, tanto para el clustering como para la detecci√≥n de anomal√≠as, los datos son preprocesados con las funciones `clean_data` y `scale` de la clase `Profile`, donde el escalamiento de las variables num√©ricas se realiza mediante un escalador logar√≠tmico y un `MinMaxScaler`, mientras que el escalamiento de las variables categ√≥ricas mediante `OneHotEncoder`.

Por un lado, durante el proceso de creaci√≥n de clusters con la funci√≥n `make_clusters` de la clase `Profile`, se considera el algoritmo de clasificaci√≥n no supervisada `KMeans` debido a su simplicidad para ser implementado. En cuanto a los hiperpar√°metros del m√©todo, se utilizaron aquellos por *default* para facilitar la elecci√≥n de la cantidad √≥ptima de clusters (`n_clusters`), la cual es determinada autom√°ticamente a trav√©s del m√©todo del codo como se observa en el bloque de c√≥digo siguiente.

Cabe aclarar que la metodolog√≠a implementada en este trabajo para determinar la cantidad √≥ptima de clusters se sustenta en utilizar un valor l√≠mite (`threshold`) para la diferencia relativa entre la suma de errores cuadrados (SSE, por sus siglas en ingl√©s) normalizada por el SSE inicial. As√≠, al fijar un valor l√≠mite del 5 % (`threshold=.05`), se obtiene que el codo se ubicar√≠a en `n_clusters=7`, cuya proyecci√≥n bidimensional, luego de reducir la dimensionalidad del subconjunto del Dataset con el algoritmo t-SNE, se muestra a continuaci√≥n.

```python
df_clusters = Profiler(df_olimpiadas_subset).make_clusters(include='all', select_dtypes=True, main_delimiter='?', 
                                                           strategy='drop', 
                                                           cat_preprocess=[OneHotEncoder], 
                                                           num_preprocess=[np.log10, MinMaxScaler],
                                                           random_state=42, make_clusters_method=[KMeans], n_clusters=None, 
                                                           threshold=.05)
```

Por otro lado, durante el proceso de detecci√≥n de anomal√≠as con la funci√≥n `detect_anomalies` de la clase `Profile`, se considera el algoritmo de clasificaci√≥n no supervisada `DBSCAN` debido a su resistencia al ruido.
```python
df_anomalies = Profiler(df_olimpiadas_subset).detect_anomalies(include='all', select_dtypes=True, main_delimiter='?', 
                                                               strategy='drop', 
                                                               cat_preprocess=[OneHotEncoder], 
                                                               num_preprocess=[np.log10, MinMaxScaler],
                                                               random_state=42, detect_anomalies_method=[DBSCAN])
```
