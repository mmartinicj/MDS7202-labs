### Tabla de Resumen

Para generar la tabla de resumen:

- [X] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [X] Hacer una simplificación a través de un mapeo (investigar el método `map()`) de la variable `parental level of education`  según la siguiente conversión: 

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [X] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas. 
- [X] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [X] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'. 


Utilizar la tabla de resultados esperados como guía para desarrollar este punto.

````python
#calcular promedio y añadir como columna de nombre 'GPA'
df_grades['GPA'] = df_grades[['math score', 'reading score', 'writing score']].mean(axis=1)
````

````python
education_map = {
    'some high school': 'school',
    'some college': 'school',
    'high school': 'school',
    'bachelor\'s degree': 'college',
    'associate\'s degree': 'college',
    'master\'s degree': 'postgraduate'
}

df_grades['simple parental level of education'] = df_grades['parental level of education'].map(education_map)
````

````python
#fija número de decimales en 2
pd.set_option("display.precision", 2)

# crea tabla agregada
summary_table = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).agg({
    'math score': 'mean',
    'reading score': 'mean',
    'writing score': 'mean',
    'GPA': 'mean'
    }).reset_index()

#añade count para cada fila
summary_table['count'] = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).size().reset_index(name='Count')['Count']

#calcula porcentaje respecto al total
total_rows = len(df_grades)
summary_table['Percentage'] = (summary_table['count'] / total_rows * 100).apply(lambda x: f"{x:.2f}%")

last_race = ""
for i, row in summary_table.iterrows():
    if row['race/ethnicity'] == last_race:
        summary_table.at[i, 'race/ethnicity'] = ''
    else:
        last_race = row['race/ethnicity']

````

````python
#Cargar other_grades como df_grades_new
df_grades_new = pd.read_csv('Lab_05/other_grades.csv').apply(pd.to_numeric, errors='ignore')
df_grades_new
````

````python
#outer join
df_outer = pd.merge(
    left=df_grades,
    right=df_grades_new,
    on='names',
    how="outer",
    sort=True,
)
print(f'Número total de filas con outer join: {df_outer.shape[0]}')
print(f'Número de nombres únicos: {df_outer["names"].nunique()}')
df_outer
````

````python
#left join
df_left = pd.merge(
    left=df_grades,
    right=df_grades_new,
    on='names',
    how="left",
    sort=True,
)
print(f'Número total de filas con left join: {df_left.shape[0]}')
print(f'Número de nombres únicos: {df_left["names"].nunique()}')

df_left
````

````python
#right join
df_right = pd.merge(
    left=df_grades,
    right=df_grades_new,
    on='names',
    how="right",
    sort=True,
)
print(f'Número total de filas con right join: {df_right.shape[0]}')
print(f'Número de nombres únicos: {df_right["names"].nunique()}')

df_right
````

````python
#inner join
df_inner = pd.merge(
    left=df_grades,
    right=df_grades_new,
    on='names',
    how="inner",
    sort=True,
)
print(f'Número total de filas con inner join: {df_inner.shape[0]}')
print(f'Número de nombres únicos: {df_inner["names"].nunique()}')

df_inner
````

> **Respuesta**: 

En este caso, el dataframe original contiene 875 filas, mientras que el dataframe nuevo (no limpiado) tiene 1000 casos. El hecho de que el número de filas del outer join sea igual al número de filas del dataframe nuevo habla de que todos los estudiantes del dataframe original están en el dataframe nuevo. Esto explica que el resultado del outer join sea el mismo que el del right join y que el resultado del inner join sea el mismo que el del left join.

Para casos como este conviene usar left join, donde a los casos del dataframe de la izquierda se les añade la información nueva que aporta el dataframe de la derecha, pero sin agregar nuevos casos contenidos en el dataframe de la derecha. Esto da como resultado el dataframe original (izquierda) con la información actualizada encontrada en el dataframe de la derecha.
